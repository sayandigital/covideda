```python
karnataka_merge=karnatak_covid.merge(karnatak_state,how='left', on=['Year','Month','DayOfMonth'])
```

5.1 What are the total samples tested, how many positive cases were reported, how many was confirmed and how many died in each day for August, Sep and oct and writing back to excel



```python
karnataka_merge.head()
```

```python
summary=karnataka_merge.groupby(['Month','DayOfMonth'])[["TotalSamples", "Positive", "Confirmed", "Deaths"]].sum()
```
